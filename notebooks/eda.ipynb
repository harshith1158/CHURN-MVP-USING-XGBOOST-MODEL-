```python
import pandas as pd
import numpy as np

np.random.seed(42)

N = 1000
df = pd.DataFrame({
    "last_login_days": np.random.randint(0, 60, N),
    "usage_score": np.random.uniform(0, 1, N),
    "tickets_last_30d": np.random.poisson(2, N),
    "avg_ticket_sentiment": np.random.uniform(-1, 1, N),
    "email_open_rate": np.random.uniform(0, 1, N),
    "feature_drop_count": np.random.poisson(1, N),
    "payment_delay_days": np.random.randint(0, 30, N),
})

# Calculate churn probability based on weighted features
df["churn_probability"] = 1 / (1 + np.exp(
    -(
        -1.5
        + 0.08 * df["last_login_days"]
        - 2.0 * df["usage_score"]
        + 0.5 * df["tickets_last_30d"]
        - 1.0 * df["avg_ticket_sentiment"]
        + 0.7 * df["feature_drop_count"]
        + 0.05 * df["payment_delay_days"]
    )
))

# Simulate churned label
df["churned"] = (df["churn_probability"] > 0.5).astype(int)

df.to_csv("../data/synthetic_churn_dataset.csv", index=False)
print("Synthetic dataset saved to data/synthetic_churn_dataset.csv")
df.head()
```
